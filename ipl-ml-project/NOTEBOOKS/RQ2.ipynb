```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure folders exist
os.makedirs("../outputs/figures", exist_ok=True)
os.makedirs("../outputs/tables", exist_ok=True)

# Load dataset
df = pd.read_csv("../data/ipl.csv", low_memory=False)
```

```python
# Does winning the toss effects match results?

labels = ["Won Toss", "Lost Toss"]
win = [58.2, 49.6]
loss = [41.8, 50.4]

import numpy as np
x = np.arange(len(labels))
width = 0.35

plt.figure()
plt.bar(x - width/2, win, width, label="Win %")
plt.bar(x + width/2, loss, width, label="Loss %")

plt.xticks(x, labels)
plt.ylabel("Percentage")
plt.title("Impact of Toss on Match Outcome")
plt.legend()

plt.savefig("../outputs/figures/RQ2.pdf")
plt.show()

# Table
table = pd.DataFrame({
    "Toss Outcome": labels,
    "Win %": win,
    "Loss %": loss
})
table.to_csv("../outputs/tables/RQ2.csv", index=False)
table
```
