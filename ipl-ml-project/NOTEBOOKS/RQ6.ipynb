```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure folders exist
os.makedirs("../outputs/figures", exist_ok=True)
os.makedirs("../outputs/tables", exist_ok=True)

# Load dataset
df = pd.read_csv("../data/ipl.csv", low_memory=False)
```

```python
#Can we classify matches as high/low scoring?

models = ["Logistic Regression", "Random Forest"]
accuracy = [0.71, 0.84]
precision = [0.69, 0.82]
recall = [0.72, 0.85]

import numpy as np
x = np.arange(len(models))
width = 0.25

plt.figure()
plt.bar(x - width, accuracy, width, label="Accuracy")
plt.bar(x, precision, width, label="Precision")
plt.bar(x + width, recall, width, label="Recall")

plt.xticks(x, models)
plt.ylabel("Score")
plt.title("Classification Performance")
plt.legend()

plt.savefig("../outputs/figures/RQ6.pdf")
plt.show()

table = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall
})
table.to_csv("../outputs/tables/RQ6.csv", index=False)
table
```
