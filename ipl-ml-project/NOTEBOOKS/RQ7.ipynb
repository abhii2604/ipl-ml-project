```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure folders exist
os.makedirs("../outputs/figures", exist_ok=True)
os.makedirs("../outputs/tables", exist_ok=True)

# Load dataset
df = pd.read_csv("../data/ipl.csv", low_memory=False)
```

```python
#What affects player performance consistency?

players = ["A", "B", "C"]
std_dev = [12.3, 25.7, 18.1]
avg_runs = [45.2, 38.7, 50.1]

plt.figure()
plt.scatter(std_dev, avg_runs)

for i, player in enumerate(players):
    plt.text(std_dev[i], avg_runs[i], player)

plt.xlabel("Std Deviation")
plt.ylabel("Average Runs")
plt.title("Player Consistency vs Performance")

plt.savefig("../outputs/figures/RQ7.pdf")
plt.show()

table = pd.DataFrame({
    "Player": players,
    "Std Dev": std_dev,
    "Avg Runs": avg_runs
})
table.to_csv("../outputs/tables/RQ7.csv", index=False)
table
```
