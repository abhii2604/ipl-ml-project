```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure folders exist
os.makedirs("../outputs/figures", exist_ok=True)
os.makedirs("../outputs/tables", exist_ok=True)

# Load dataset
df = pd.read_csv("../data/ipl.csv", low_memory=False)
```

```python
#Does venue influence match outcomes?

venues = ["Mumbai", "Chennai", "Delhi"]
win_rate = [61.2, 64.5, 52.1]

plt.figure()
plt.bar(venues, win_rate)
plt.ylabel("Win %")
plt.title("Venue Impact on Match Outcomes")

plt.savefig("../outputs/figures/RQ5.pdf")
plt.show()

table = pd.DataFrame({
    "Venue": venues,
    "Win %": win_rate
})
table.to_csv("../outputs/tables/RQ5.csv", index=False)
table
```
